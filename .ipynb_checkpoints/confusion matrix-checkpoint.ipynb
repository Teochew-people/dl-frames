```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bls = np.array([[5, 3, 1, 1, 0],
       [3, 4, 2, 0, 3],
       [1, 3, 4, 0, 0],
       [1, 3, 3, 4, 2],
       [1, 3, 0, 1, 4]])
```

```python
ann = np.array([[0, 1, 3, 1, 0],
              [2, 3, 0, 0, 1],
              [0, 3, 2, 0, 0],
              [0, 4, 3, 0, 2],
              [2, 0, 1, 2, 4]])
```

```python
cnn = np.array([[0, 2, 1, 0, 3],
       [0, 2, 2, 0, 4],
       [0, 2, 0, 2, 4],
       [0, 4, 0, 2, 3],
       [0, 1, 1, 2, 7]])
```

```python
lstm = np.array([[1, 3, 3, 1, 3],
       [4, 2, 1, 3, 1],
       [1, 2, 2, 2, 3],
       [4, 4, 2, 4, 1],
       [1, 2, 0, 2, 5]])
```

```python
x = ['P1', 'P2', 'P3', 'P4', 'P5']
y = ['T1', 'T2', 'T3', 'T4', 'T5']

pic = sns.heatmap(ann,annot=True,yticklabels=y, linewidths=0.5, xticklabels=x,cmap="YlGnBu")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
sns.set(font_scale=1.5)
pic_save = pic.get_figure()
pic_save.savefig('./pictures/ann-cfm.eps')
plt.savefig('../../conference mauscript/Conference-Chai/latex/figures/ann-cfm.eps')
pic
```

```python
pic = sns.heatmap(cnn,annot=True,yticklabels=y, linewidths=0.5, xticklabels=x,cmap="YlGnBu")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
sns.set(font_scale=1.5)
pic_save = pic.get_figure()
pic_save.savefig('./pictures/cnn-cfm.eps')
plt.savefig('../../conference mauscript/Conference-Chai/latex/figures/cnn-cfm.eps')
pic
```

```python
pic = sns.heatmap(lstm,annot=True,yticklabels=y, linewidths=0.5, xticklabels=x,cmap="YlGnBu")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
sns.set(font_scale=1.5)
pic_save = pic.get_figure()
pic_save.savefig('./pictures/lstm-cfm.eps')
plt.savefig('../../conference mauscript/Conference-Chai/latex/figures/lstm-cfm.eps')
pic
```

```python
pic = sns.heatmap(bls,annot=True,yticklabels=y, linewidths=0.5, xticklabels=x,cmap="YlGnBu")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
sns.set(font_scale=1.5)
pic_save = pic.get_figure()
pic_save.savefig('./pictures/bls-cfm.eps')
plt.savefig('../../conference mauscript/Conference-Chai/latex/figures/bls-cfm.eps')
pic
```
